```python
#Tabular Confusion Matrix (Grading)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import numpy as np

data = pd.read_csv('/kaggle/input/egg-figure-paper-generate/Grading_Tabular_AdaBoost_fold_predictions.csv')

true_labels = data["True_Label"]
predicted_labels = data['Predicted_Label']

# Compute confusion matrix
cm = confusion_matrix(true_labels, predicted_labels, labels=['High', 'Low'])

# Normalize the confusion matrix to percentages
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100

# Create a DataFrame for better labeling
cm_df = pd.DataFrame(cm_normalized, index=['High', 'Low'], columns=['High', 'Low'])

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_df, annot=False, fmt='.2f', cmap='Blues', cbar_kws={'label': 'Percentage'}, annot_kws={"size": 12})
plt.ylabel('Actual', fontsize=11)
plt.xlabel('Predicted', fontsize=11)
plt.tight_layout()

# Add count annotations and percentage sign
for i in range(len(cm)):
    for j in range(len(cm[i])):
        if(i==0 and j==0) or (i==1 and j==1):
            plt.text(j + 0.5, i + 0.60, f'{cm_normalized[i][j]:.2f}%\n({cm[i][j]})', ha='center', va='center', color='white', fontsize=12)
        else:
            plt.text(j + 0.5, i + 0.60, f'{cm_normalized[i][j]:.2f}%\n({cm[i][j]})', ha='center', va='center', color='black', fontsize=12)

plt.savefig("Grading_Tabular_Confusion_Matrix_Figure.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
#ROC - AUC Curve (Grading)

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
import numpy as np

# Load the three CSV files
tabular_data = pd.read_csv('/kaggle/input/egg-figure-paper-generate/Grading_Tabular_AdaBoost_fold_predictions.csv')
image_data = pd.read_csv('/kaggle/input/egg-figure-paper-generate/grading_imageOnly_ensemble_predictions.csv')
multimodal_data = pd.read_csv('/kaggle/input/egg-figure-paper-generate/grading_merged_ensemble_predictions.csv')

# Assuming each CSV has 'true_label' (0 or 1), 'predicted_label', and 'prediction_probability' columns
# Convert true labels to binary (0 or 1) if they are strings like 'High'/'Low'
tabular_true = (tabular_data['True_Label'].map({'High': 0, 'Low': 1}) if tabular_data['True_Label'].dtype == 'object' else tabular_data['True_Label']).astype(int)
image_true = (image_data['True_Label'].map({'High': 0, 'Low': 1}) if image_data['True_Label'].dtype == 'object' else image_data['True_Label']).astype(int)
multimodal_true = (multimodal_data['True_Label'].map({'High': 0, 'Low': 1}) if multimodal_data['True_Label'].dtype == 'object' else multimodal_data['True_Label']).astype(int)

tabular_prob = tabular_data['Class_Probability']
image_prob = image_data['Ensemble_Class_Probability']
multimodal_prob = multimodal_data['Ensemble_Class_Probability']

# Compute ROC curve and ROC area for each case
fpr_tabular, tpr_tabular, _ = roc_curve(tabular_true, tabular_prob)
roc_auc_tabular = auc(fpr_tabular, tpr_tabular)

fpr_image, tpr_image, _ = roc_curve(image_true, image_prob)
roc_auc_image = auc(fpr_image, tpr_image)

fpr_multimodal, tpr_multimodal, _ = roc_curve(multimodal_true, multimodal_prob)
roc_auc_multimodal = auc(fpr_multimodal, tpr_multimodal)

# Plot all ROC curves
plt.figure(figsize=(8, 6))
plt.plot(fpr_tabular, tpr_tabular, color='blue', lw=2, label=f'Tabular (AUC = {roc_auc_tabular:.2f})')
plt.plot(fpr_image, tpr_image, color='red', lw=2, label=f'ImageOnly Ensemble (AUC = {roc_auc_image:.2f})')
plt.plot(fpr_multimodal, tpr_multimodal, color='green', lw=2, label=f'Multimodal Ensemble (AUC = {roc_auc_multimodal:.2f})')

# Plot the diagonal line (random guess)
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')

# Customize the plot
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.savefig("Grading_ROC_AUC_CURVE_Figure.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
